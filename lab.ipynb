```python
column_names = file1.columns
```

```python
data = pd.DataFrame(columns = column_names)
```

```python
data = pd.concat([data,file1,file2,file3], axis = 0)
```

```python
data.head()
```

```python
data['GENDER'] = np.where(data['GENDER'].isna(),data['Gender'],data['GENDER'])
data['State'] = np.where(data['State'].isna(),data['ST'],data['State'])

data = data.drop(['Gender', 'ST'],axis=1)
data.head()
```

```python
cols = []
for i in range(len(data.columns)):
    cols.append(data.columns[i].lower().replace(" ","_"))
data.columns = cols
```

```python
data.drop(['education','number_of_open_complaints'], axis = 1)
```

```python
data['customer_lifetime_value'] = data['customer_lifetime_value'].str.replace("%","")
data['customer_lifetime_value'] = data['customer_lifetime_value'].astype(float)
data.dtypes
```
